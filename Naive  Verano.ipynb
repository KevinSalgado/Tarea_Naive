**Teorema de Bayes**

Nos dice con qu frecuencia ocurre A dado que ocurre B, escrito como P(A|B), cuando sabemos con qu frecuencia ocurre B dado que ocurre A, escrito como P(B|A), y cun probables son A y B por separado.

P(A|B) es "Probabilidad de A dado B", la probabilidad de A dado que ocurre B.
P(A) es la probabilidad de A.
P(B|A) es "Probabilidad de B dado A", la probabilidad de B dado que ocurre A.
P(B) es la probabilidad de B.
Cuando P(Fuego) significa con qu frecuencia hay fuego, y P(Humo) significa con qu frecuencia vemos humo, entonces:

P(Fuego|Humo) significa con qu frecuencia hay fuego cuando vemos humo.
P(Humo|Fuego) significa con qu frecuencia vemos humo cuando hay fuego.

Entonces, la frmula nos dice como ver el futuro viendo "hacia adelante" cuando conocemos como se vio el pasado viendo "hacia atrs"

Ejemplo: Si los incendios peligrosos son raros (1%) pero el humo es bastante comn (10%) debido a las fbricas, y el 90% de los incendios peligrosos producen humo, entonces:

P(Fuego|Humo) = P(Fuego) P(Humo|Fuego) = 1% x 90% = 9%P(Humo)10%

En este caso, el 9% del tiempo se espera que el humo signifique un incendio peligroso.

# Clasificador de Naive Bayes (DEFINICION)
El clasificador de Naive Bayes calcula las probabilidades para cada factor (en este caso del ejemplo de correo electrnico seran Alicia y Roberto para una caracterstica de entrada determinada). Luego selecciona el resultado con la probabilidad ms alta.

Este clasificador asume que las caractersticas (en este caso, las palabras que hemos usado como entrada) son independientes. De ah el trmino "naive" (ingenuo). A pesar de esto, es un algoritmo potente utilizado para:

Prediccin en tiempo real
Clasificacin de texto / Filtrado de spam
Sistemas de recomendacin
Entonces, matemticamente podemos escribirlo de la siguiente manera:

Si tenemos un evento E y actores de prueba x1, x2, x3, etc.

Primero calculamos P(x1|E), P(x2|E)... [ledo como probabilidad de x1 dado que ocurri el evento E] y luego seleccionamos el actor de prueba x con el valor de probabilidad mxima.

```python
import os
import numpy as np
from collections import Counter
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

def make_Dictionary(root_dir):
    all_words = []
    emails = [os.path.join(root_dir, f) for f in os.listdir(root_dir)]
    for mail in emails:
        with open(mail) as m:
            for line in m:
                words = line.split()
                all_words += words
    dictionary = Counter(all_words)
    list_to_remove = list(dictionary.keys())
    for item in list_to_remove:
        if not item.isalpha() or len(item) == 1:
            del dictionary[item]
    dictionary = dictionary.most_common(3000)
    return dictionary

def extract_features(mail_dir):
    files = [os.path.join(mail_dir, fi) for fi in os.listdir(mail_dir)]
    features_matrix = np.zeros((len(files), 3000))
    train_labels = np.zeros(len(files))
    docID = 0
    for fil in files:
        with open(fil) as fi:
            for i, line in enumerate(fi):
                if i == 2:
                    words = line.split()
                    for word in words:
                        for i, d in enumerate(dictionary):
                            if d[0] == word:
                                wordID = i
                                features_matrix[docID, wordID] = words.count(word)
        train_labels[docID] = 0
        filepathTokens = fil.split('/')
        lastToken = filepathTokens[-1]
        if lastToken.startswith("spmsg"):
            train_labels[docID] = 1
        docID += 1
    return features_matrix, train_labels

TRAIN_DIR = "../train-mails"
TEST_DIR = "../test-mails"

dictionary = make_Dictionary(TRAIN_DIR)
print("Reading and processing emails from file.")
features_matrix, labels = extract_features(TRAIN_DIR)
test_feature_matrix, test_labels = extract_features(TEST_DIR)

model = GaussianNB()
print("Training model.")
model.fit(features_matrix, labels)

predicted_labels = model.predict(test_feature_matrix)
accuracy = accuracy_score(test_labels, predicted_labels)
print("FINISHED classifying. Accuracy score: {:.2%}".format(accuracy))
```
